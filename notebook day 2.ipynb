```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = [10,12,13,12,11,200,14,13,15,12]

sns.boxplot(data)
plt.show()
```

```python
Q1,Q3 = np.percentile(data,[25,75])
IQR = Q3-Q1
lower,upper = Q1-1.5* IQR, Q3 + 1.5* IQR
outliers = [x for x in data if x < lower or x > upper]

print("Outliers detected:",outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.DataFrame({'Age': [18,22,30,40],
                   'Salary': [20000,50000,80000,120000]})

scaler = MinMaxScaler()
df_norm = scaler.fit_transform(df)

scaler_std = scaler_std = StandardScaler()
df_std = scaler_std.fit_transform(df)

print("Original Data:\n",df)
print("\nNormalized Data:\n",df_norm)
print("\nStandardized Data:\n",df_std)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie'],
    'Age': [25,32,47],
    'Country': ['US','India','US'],
    'Spend1': [200,150,300],
    'Spend2': [100,200,400]
})

df['Total_Spend'] = df['Spend1'] + df['Spend2']

df['Age_Group'] = pd.cut(df['Age'],bins=[0,30,50],labels=['Young','Middle'])

encoder = LabelEncoder()
df['County_encoded'] = encoder.fit_transform(df['Country'])

print(df)
```
